```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
Salary_df = pd.read_csv("Salary_Data.csv")
Salary_df = Salary_df.dropna()
```

```python
edu_level_unique = Salary_df['Education Level'].unique()
job_title_unique = Salary_df['Job Title'].unique()
num_of_job_titles = Salary_df['Job Title'].nunique()

edu_level_unique, job_title_unique, num_of_job_titles
```

```python
clean_salary_df = pd.DataFrame({
    'Age': Salary_df['Age'].astype(int),
    'Gender': Salary_df['Gender'],
    'Education Level': Salary_df['Education Level'].replace('Bachelor\'s', 'Bachelor\'s Degree').replace('Master\'s', 'Master\'s Degree').replace('nan', '*no education level recorded*').replace('phD', 'PhD'),
    'Job Title': Salary_df['Job Title'].replace('Customer Service Rep', 'Customer Service Representative').replace('Front end Developer', 'Front End Developer').replace('IT Support Specialist', 'IT Support').replace('Social M', 'Social Media').replace('Social Media Man', 'Social Media Manager'),
    'Years of Experience': Salary_df['Years of Experience'].astype(int),
    'Salary': Salary_df['Salary'].astype(int)
})

clean_salary_df
```

```python
clean_salary_df['Job Title'].nunique()
```

```python
for x in clean_salary_df:
    print(f"{x}:")
    print(clean_salary_df[f'{x}'].unique())
    print('\n\n')
```

```python
for x in clean_salary_df:
    print(f"{x}:")
    print(clean_salary_df[clean_salary_df[f'{x}'] == np.nan])
    print('\n\n')
```

```python
clean_salary_df.to_csv('cleaned_salary.csv', index=False)
```
