```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
Salary_df = pd.read_csv("Salary_Data.csv")
Salary_df = Salary_df.dropna()
Salary_df.dtypes
```

```python
# Salary_df['Education Level'].unique()
# Salary_df['Job Title'].unique()
Salary_df['Job Title'].nunique()
```

```python
# Salary_df[['Age'] ]= Salary_df[['Age']].apply(pd.to_numeric)
# Salary_df['Age']


Salary_df['Age'] = Salary_df['Age'].astype(int)
Salary_df['Age']
```

```python
clean_salary_df = pd.DataFrame({
    'Age': Salary_df['Age'].astype(int),
    'Gender': Salary_df['Gender'],
    'Education Level': Salary_df['Education Level'].replace('Bachelor\'s', 'Bachelor\'s Degree').replace('Master\'s', 'Master\'s Degree').replace('nan', '*no education level recorded*'),
    'Job Title': Salary_df['Job Title'].replace('Customer Service Rep', 'Customer Service Representative').replace('Front end Developer', 'Front End Developer').replace('IT Support Specialist', 'IT Support').replace('Social M', 'Social Media').replace('Social Media Man', 'Social Media Manager'),
    'Years of Experience': Salary_df['Years of Experience'].astype(int),
    'Salary': Salary_df['Salary'].astype(int)
})

clean_salary_df.head()
```
